```python
def getBestParams(auc_table, i = 0):
    target_auc = np.max(auc_table.iloc[:,0])
    sort = sorted(zip(np.abs(auc_table.iloc[:,0] - target_auc), range(auc_table.shape[0])))
    return(auc_table.loc[sort[i][1], :])
```

```python
def getOneStandParams(auc_table, i = 0):
    target_auc = np.mean(auc_table[:,0]) + np.std(auc_table[:,0])
    sort = sorted(zip(np.abs(auc_table[:,0] - target_auc), range(auc_table.shape[0])))
    return(auc_table[sort[i][1], :])
```
